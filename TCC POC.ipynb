```python
#Import dos módulos
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
#Retrieve/engineering no dataset do S&P
esp = yf.Ticker("^GSPC")
espDf = esp.history(period = "max")
espDf = espDf.sort_values("Date", ascending = True)
espDf.reset_index(inplace = True)
espDf = espDf[~(espDf['Date'] < '2003-12-01')]
espDf['Date'] = espDf['Date'].values.astype('<M8[D]')
espDf['timestamp'] = pd.to_datetime(espDf.Date).dt.tz_localize(None)
espDf.drop(['Dividends', 'Stock Splits', 'timestamp'], axis = 1, inplace = True)
espDf
```

```python
#Retrieve/engineering no dataset do dolar
dol = yf.Ticker("USDBRL=X")
dolDf = dol.history(period = "max")
dolDf = dolDf.sort_values("Date", ascending = True)
dolDf.reset_index(inplace = True)
espDf['Date'] = espDf['Date'].values.astype('<M8[D]')
dolDf = dolDf[~(dolDf['Date'] < '2003-12-01')]
dolDf['Date'] = pd.to_datetime(dolDf.Date).dt.tz_localize(None)
dolDf.drop(['Dividends', 'Stock Splits','Volume', 'Open', 'High', 'Low'], axis = 1, inplace = True)
dolDf
```

```python
#dataframe merge
espDf = espDf.merge(dolDf, on = 'Date', how = 'inner')
```

```python
espDf['rolling_close'] = espDf['Close_x'].rolling(3).mean()
espDf
```
